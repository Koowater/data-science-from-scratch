```python
from matplotlib import pyplot as plt
import random
%matplotlib notebook
```

```python
x = []
for _ in range(1, 51): x.append(random.random())
for i in x:
    plt.axvline(i)
ax = plt.gca()
ax.axes.yaxis.set_visible(False)
plt.show()
```

```python
x = []
y = []
for _ in range(1, 51):
    x.append(random.random())
    y.append(random.random())
plt.scatter(x, y, s=10)
plt.show()
```

```python
x = []
y = []
z = []
for _ in range(1, 51):
    x.append(random.random())
    y.append(random.random())
    z.append(random.random())

ax = plt.gca(projection='3d')
ax.scatter(x, y, z)
plt.show()
```
